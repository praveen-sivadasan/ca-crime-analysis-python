# Question 5
-Use Gutenberg and Web_text data. Find out what are the top 5 words that Shakespeare used but we do not use in currently.
-Take top 50 words from Shakespeare (all 3 books) and top 50 from Web_text (all the records).
-Remove punctuation and stop words.
-Remove the words we still use today, and get the unused list. Show the top 5 elements. 


```python
import nltk
from nltk.corpus import gutenberg
from nltk.corpus import webtext
from nltk.corpus import stopwords

gfiles = gutenberg.fileids()
wfiles = webtext.fileids()
gwords={}
ban_list = stopwords.words('english')
```

```python
#gfiles
```

```python
web_words={}
for file in wfiles:
    word_data = webtext.words(file)
    filtered_words = [word for word in word_data if word not in ban_list]
    for word in filtered_words:
        if word.isalpha():
            web_words[word] = web_words.get(word, 0) + 1

'''       
web_words.pop('.')
web_words.pop(':')
web_words.pop(';')
web_words.pop(',')
web_words.pop("'")
web_words.pop('?')
web_words.pop('!')
web_words.pop('-')
web_words.pop('[')
web_words.pop(']')
web_words.pop(')')
web_words.pop('(')
web_words.pop('"')
'''

#for key, value in gwords.items():
    #print(key,'-',value)
#print('**** web text data \n')
#print(web_words)
    
sorted_web_words = [(k, web_words[k]) for k in sorted(web_words, key=web_words.get, reverse=True)]
#import operator
#orted_web_words = sorted(web_words.items(), key=operator.itemgetter(1))


#print('***Sorted web text data \n')
#print(sorted_web_words)


count = 0 
lim_wDic = {}
for k, v in sorted_web_words:
    if 1==1:#count < 50:
        #print(k,'-',v)
        lim_wDic[k] = v;
        count += 1
    else:
        break
            
```

```python
#filtered_words = [word for word in word_list if word not in stopwords.words('english')]

for file in gfiles:
    if 'shakespeare' in file:
        word_data = gutenberg.words(file)
        #seperating words from stopwords
        filtered_words = [word for word in word_data if word not in ban_list]
        for word in filtered_words:
            if word.isalpha():
                gwords[word] = gwords.get(word, 0) + 1        
                

'''
gwords.pop(',')
gwords.pop("'")
gwords.pop('?')
gwords.pop('!')
gwords.pop('-')
gwords.pop('[')
gwords.pop(']')
gwords.pop(')')
gwords.pop('(')
gwords.pop('"')
'''

#print('Gutenberg data post stop wrods and excalamation removal')
#print(gwords)

#for key, value in gwords.items():    #print(key,'-',value)
sorted_gwords = [(k, gwords[k]) for k in sorted(gwords, key=gwords.get, reverse=True)]

#print('Gutenberg data post sorting')
#print(sorted_gwords)


#print('***Top 50 \n')
count = 0 
lim_gDic = {}
for k, v in sorted_gwords:
    if 1==1:#count < 50:
        #print(k,'-',v)
        lim_gDic[k] = v;
        count += 1
    else:
        break
            
            
```
